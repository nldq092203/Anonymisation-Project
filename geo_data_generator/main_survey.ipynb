```python
from datetime import datetime
from models.survey import Survey
from utils.format import replace_commas_with_tabs_in_place
```

```python
# Define survey parameters
city_name = "Paris"  # City name to dynamically determine center point
radius = 10000  # 10 km radius
number_of_people = 1
start_date = datetime(2024, 10, 1)  # Start of the survey
end_date = datetime(2024, 11, 1)  # End of the survey
```

```python
# Create the survey
survey = Survey(city_name, radius, number_of_people, start_date, end_date)
```

```python
survey_data = survey.simulate(max_records_per_day=3)   
output_file = "origin_data/bigger_survey_results.csv"
survey.save_to_csv(survey_data, output_file)
replace_commas_with_tabs_in_place(output_file)
print("Survey completed and results saved to 'big_survey_results.csv'.")
```
